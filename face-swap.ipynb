# Extracting the First Index from a NumPy Array

- The function `extract_index_nparray` takes a NumPy array (typically a 2D array).
- It accesses the first sub-array (`nparray[0]`).
- Iterates through this sub-array and returns the first element.
- Useful when you want to extract the first index from results like those returned by `np.where`.


```python
# Extracting index from array
def extract_index_nparray(nparray):
    index = None
    for num in nparray[0]:
        index = num
        break
    return index
```

```python
img = np.array(image1)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
mask = np.zeros_like(img_gray)
img2 = np.array(image2)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

# Converting Images to Grayscale and Initializing Mask

- Converts `image1` from PIL Image to NumPy array for OpenCV processing.
- Changes the color space from BGR to Grayscale using OpenCV’s `cvtColor`.
- Creates a mask array of zeros with the same shape as the grayscale image; useful for drawing or segmentation.
- Similarly converts `image2` to NumPy array and grayscale for further processing.


```python
# Initalizing frontal face detector and shape predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("/kaggle/working/shape_predictor_68_face_landmarks.dat")
height, width, channels = img2.shape
img2_new_face = np.zeros((height, width, channels), np.uint8)
```

# Facial Landmark Extraction and Delaunay Triangulation

- Detects faces in the grayscale image using a face detector (`detector`).
- For each detected face:
  - Uses a landmark predictor (`predictor`) to find 68 facial landmark points.
  - Stores the (x, y) coordinates of these landmarks in a list.
  - Converts the list to a NumPy array for OpenCV operations.
  - Computes the convex hull around the landmarks to outline the face.
  - Fills the convex hull area in a mask to isolate the face region.
  - Applies bitwise AND to extract the face region from the original image using the mask.
  - Calculates a bounding rectangle around the convex hull.
  - Initializes OpenCV’s `Subdiv2D` for Delaunay triangulation within the bounding rectangle.
  - Inserts the landmark points into the subdivision.
  - Retrieves triangles formed by the subdivision.
  - For each triangle, finds the indices of the triangle’s vertices corresponding to the original landmarks.
  - Collects these indices in `indexes_triangles` for further processing like warping or morphing.


```python
# Face 1
faces = detector(img_gray)
for face in faces:
    landmarks = predictor(img_gray, face)
    landmarks_points = []
    for n in range(0, 68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        landmarks_points.append((x, y))

    points = np.array(landmarks_points, np.int32)
    convexhull = cv2.convexHull(points)
    cv2.fillConvexPoly(mask, convexhull, 255)

    face_image_1 = cv2.bitwise_and(img, img, mask=mask)

    # Delaunay triangulation
    rect = cv2.boundingRect(convexhull)
    subdiv = cv2.Subdiv2D(rect)
    subdiv.insert(landmarks_points)
    triangles = subdiv.getTriangleList()
    triangles = np.array(triangles, dtype=np.int32)

    indexes_triangles = []
    for t in triangles:
        pt1 = (t[0], t[1])
        pt2 = (t[2], t[3])
        pt3 = (t[4], t[5])


        index_pt1 = np.where((points == pt1).all(axis=1))
        index_pt1 = extract_index_nparray(index_pt1)

        index_pt2 = np.where((points == pt2).all(axis=1))
        index_pt2 = extract_index_nparray(index_pt2)

        index_pt3 = np.where((points == pt3).all(axis=1))
        index_pt3 = extract_index_nparray(index_pt3)

        if index_pt1 is not None and index_pt2 is not None and index_pt3 is not None:
            triangle = [index_pt1, index_pt2, index_pt3]
            indexes_triangles.append(triangle)
```

# Facial Landmark Detection and Convex Hull for Second Image

- Detects faces in the grayscale second image (`img2_gray`) using the face detector.
- For each detected face:
  - Uses the landmark predictor to extract 68 facial landmark points.
  - Collects the (x, y) coordinates of these landmarks in a list.
  - Converts this list into a NumPy array for processing.
  - Calculates the convex hull around the landmarks to get the outline of the face region.


```python
# Face 2
faces2 = detector(img2_gray)
for face in faces2:
    landmarks = predictor(img2_gray, face)
    landmarks_points2 = []
    for n in range(0, 68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        landmarks_points2.append((x, y))


    points2 = np.array(landmarks_points2, np.int32)
    convexhull2 = cv2.convexHull(points2)
```

# Creating Empty Masks for Face Morphing

- Initializes `lines_space_mask` as a black (zero) image with the same shape as the first grayscale image; used to draw triangle edges or facial lines.
- Creates `lines_space_new_face` as a zero-filled image with the same shape as the second original image; used to accumulate the morphed face output.


```python
# Creating empty mask
lines_space_mask = np.zeros_like(img_gray)
lines_space_new_face = np.zeros_like(img2)
```

# Delaunay Triangulation and Triangle Warping for Face Morphing

- Iterates over each triangle defined by indices in `indexes_triangles`.
- For the first face:
  - Extracts the three landmark points forming the triangle.
  - Computes a bounding rectangle for the triangle.
  - Crops the triangular region from the original image.
  - Creates a mask for the triangle region to isolate it.
  - Draws triangle edges on a mask image (`lines_space_mask`).
- For the second face:
  - Extracts corresponding triangle landmark points.
  - Computes bounding rectangle and creates a mask.
- Calculates the affine transformation matrix (`M`) mapping the first triangle to the second.
- Warps the cropped triangle from the first face to match the second face's triangle shape.
- Masks and overlays the warped triangle onto the corresponding region of the new face image (`img2_new_face`).
- Uses bitwise operations and thresholding to blend triangles smoothly without overwriting existing regions.
- Reconstructs the entire morphed face by repeating this for all triangles.


```python
# Triangulation of both faces
for triangle_index in indexes_triangles:
    # Triangulation of the first face
    tr1_pt1 = landmarks_points[triangle_index[0]]
    tr1_pt2 = landmarks_points[triangle_index[1]]
    tr1_pt3 = landmarks_points[triangle_index[2]]
    triangle1 = np.array([tr1_pt1, tr1_pt2, tr1_pt3], np.int32)


    rect1 = cv2.boundingRect(triangle1)
    (x, y, w, h) = rect1
    cropped_triangle = img[y: y + h, x: x + w]
    cropped_tr1_mask = np.zeros((h, w), np.uint8)


    points = np.array([[tr1_pt1[0] - x, tr1_pt1[1] - y],
                       [tr1_pt2[0] - x, tr1_pt2[1] - y],
                       [tr1_pt3[0] - x, tr1_pt3[1] - y]], np.int32)

    cv2.fillConvexPoly(cropped_tr1_mask, points, 255)

    # Lines space
    cv2.line(lines_space_mask, tr1_pt1, tr1_pt2, 255)
    cv2.line(lines_space_mask, tr1_pt2, tr1_pt3, 255)
    cv2.line(lines_space_mask, tr1_pt1, tr1_pt3, 255)
    lines_space = cv2.bitwise_and(img, img, mask=lines_space_mask)

    # Triangulation of second face
    tr2_pt1 = landmarks_points2[triangle_index[0]]
    tr2_pt2 = landmarks_points2[triangle_index[1]]
    tr2_pt3 = landmarks_points2[triangle_index[2]]
    triangle2 = np.array([tr2_pt1, tr2_pt2, tr2_pt3], np.int32)


    rect2 = cv2.boundingRect(triangle2)
    (x, y, w, h) = rect2

    cropped_tr2_mask = np.zeros((h, w), np.uint8)

    points2 = np.array([[tr2_pt1[0] - x, tr2_pt1[1] - y],
                        [tr2_pt2[0] - x, tr2_pt2[1] - y],
                        [tr2_pt3[0] - x, tr2_pt3[1] - y]], np.int32)

    cv2.fillConvexPoly(cropped_tr2_mask, points2, 255)

    # Warp triangles
    points = np.float32(points)
    points2 = np.float32(points2)
    M = cv2.getAffineTransform(points, points2)
    warped_triangle = cv2.warpAffine(cropped_triangle, M, (w, h))
    warped_triangle = cv2.bitwise_and(warped_triangle, warped_triangle, mask=cropped_tr2_mask)

    # Reconstructing destination face
    img2_new_face_rect_area = img2_new_face[y: y + h, x: x + w]
    img2_new_face_rect_area_gray = cv2.cvtColor(img2_new_face_rect_area, cv2.COLOR_BGR2GRAY)
    _, mask_triangles_designed = cv2.threshold(img2_new_face_rect_area_gray, 1, 255, cv2.THRESH_BINARY_INV)
    warped_triangle = cv2.bitwise_and(warped_triangle, warped_triangle, mask=mask_triangles_designed)

    img2_new_face_rect_area = cv2.add(img2_new_face_rect_area, warped_triangle)
    img2_new_face[y: y + h, x: x + w] = img2_new_face_rect_area
```

# Preparing Masks for Face Swapping

- Creates a black mask (`img2_face_mask`) matching the shape of the second grayscale image.
- Fills the convex hull of the second face (`convexhull2`) with white (255) to create a mask of the face region.
- Inverts the mask to get the area **outside** the face (`img2_face_mask`), useful for blending swapped face onto the original image while preserving the background.


```python
# Face swapped (putting 1st face into 2nd face)
img2_face_mask = np.zeros_like(img2_gray)
img2_head_mask = cv2.fillConvexPoly(img2_face_mask, convexhull2, 255)
img2_face_mask = cv2.bitwise_not(img2_head_mask)
```

# Combining the Swapped Face with Original Image Background

- Uses the inverted face mask (`img2_face_mask`) to keep only the background (head without face) from the second image (`img2`).
- Applies a bitwise AND operation to extract this background portion.
- Adds the newly created swapped face (`img2_new_face`) onto the background using `cv2.add`.
- The result is a composite image where the first face is placed seamlessly onto the second image’s head region.


```python
img2_head_noface = cv2.bitwise_and(img2, img2, mask=img2_face_mask)
result = cv2.add(img2_head_noface, img2_new_face)
```

# Seamless Cloning to Blend Swapped Face

- Calculates the bounding rectangle around the second face’s convex hull (`convexhull2`).
- Determines the center point of this rectangle to serve as the cloning center.
- Uses OpenCV’s `seamlessClone` function to blend the swapped face (`result`) onto the original second image (`img2`).
- The mask `img2_head_mask` defines the region to blend.
- The `cv2.NORMAL_CLONE` flag ensures smooth, natural-looking blending without visible seams.
- The output (`seamlessclone`) is the final face-swapped image with improved realism.


```python
# Creating seamless clone of two faces
(x, y, w, h) = cv2.boundingRect(convexhull2)
center_face2 = (int((x + x + w) / 2), int((y + y + h) / 2))
seamlessclone = cv2.seamlessClone(result, img2, img2_head_mask, center_face2, cv2.NORMAL_CLONE)
```

- Converts the OpenCV image (`seamlessclone`), which is a NumPy array in BGR format, into a PIL Image for display.

- This allows inline visualization in notebooks or further image processing with PIL.
